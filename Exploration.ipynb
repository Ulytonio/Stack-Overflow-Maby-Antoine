```python
# Nous allons extraire les tags dans une liste pour pouvoir mieux les étudiers

def extrac_tags(x):
    l = []
    split_1 = x.split('<')
    for i in split_1 : 
        if i != '':
            word = i.split('>')[0]
            l.append(word)
    return l
```

```python
data['Tags_list'] = [extrac_tags(t) for t in data['Tags']]
data['Tags'] = [','.join(extrac_tags(t))for t in data['Tags']]
```

```python
# Nous allons maintenant regarder le nombre de tags et les plus utilisés

def count_split_tags(df, column, separator):
    list_words = []
    for word in df[column].str.split(separator):
        list_words.extend(word)
    df_list_words = pd.DataFrame(list_words, columns=["Tag"])
    df_list_words = df_list_words.groupby("Tag")\
        .agg(tag_count=pd.NamedAgg(column="Tag", aggfunc="count"))
    df_list_words.sort_values("tag_count", ascending=False, inplace=True)
    return df_list_words
```

```python
tags_list = count_split_tags(df=data, column='Tags', separator=',')
print("Le jeu de données compte {} tags.".format(tags_list.shape[0]))
```

```python
fig = plt.figure(figsize=(15, 8))
sns.barplot(data=tags_list.iloc[0:50, :],
            x=tags_list.iloc[0:50, :].index,
            y="tag_count", color="#f48023")
plt.xticks(rotation=90)
plt.show()
```

```python
# Nous allons ainsi nous limiter aux tags les plus populaires.

def filter_tag(x, top_list):
    temp_list = []
    for item in x:
        if (item in top_list):
            temp_list.append(item)
    return temp_list
```

```python
# J'avais dans un premier temps considérer tous les tags qui ont au moins 500 utilisations. Finalement, pour des questions
# de performances des modèles et de temps de calculs, j'ai décidé de me limiter aux 50 plus utilisés

# list_tag = tags_list.iloc[0:100].index
list_tag = tags_list.iloc[0:50].index
```

```python
data['Tags_list_bis'] = [filter_tag(t,list_tag) for t in data['Tags_list']]
```

```python
# Nous allons maintenant regarder le nombres de tags par question

data['Tags_count'] = [len(t) for t in data['Tags_list']]
data['Tags_count_bis'] = [len(t) for t in data['Tags_list_bis']]
```

# Partie 3 : Nettoyage du Corpus

```python
test = data['Body'].loc[13620]
test
```

```python
from bs4 import BeautifulSoup
test = BeautifulSoup(test,"lxml").get_text()
test
```

```python
# Appliquons BeautifulSoup à l'ensemble de notre corpus

data['Body'] = [BeautifulSoup(text,"lxml").get_text() for text in data['Body']]
data['Title'] = [BeautifulSoup(text,"lxml").get_text() for text in data['Title']]
```

```python
import re

# Nous allons maintenant utiliser Regular Expression pour supprimer certains caractéres. Comme on peut le voir dans notre 
# exemple, il reste encore des caractères à supprimer

def clean_text(x):
    # Remove unicode characters
    x = x.encode("ascii", "ignore").decode()
    # Remove English contractions
    x = re.sub("\'\w+", '', x)
    # Remove ponctuation but not # (for C# for example)
    x = re.sub('[^\\w\\s#\\s+]', '', x)
    # Remove links
    x = re.sub(r'http*\S+', '', x)
    # Remove numbers
    x = re.sub(r'\w*\d+\w*', '', x)
    # Remove extra spaces
    x = re.sub('\s+', ' ', x)
    # Case normalization
    x = x.lower()
    x = x.replace("c #", "c#")
    return x
```

```python
clean_text(test)
```

```python
data['Body'] = [clean_text(text) for text in data['Body']]
data['Title'] = [clean_text(text) for text in data['Title']]
```

```python
import nltk

# Tockenization

data['Body'] = data.Body.apply(nltk.tokenize.word_tokenize)
data['Title'] = data.Title.apply(nltk.tokenize.word_tokenize)
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')

# List of stop words in "EN" from NLTK
stop_words = stopwords.words("english")

# Remove stop words
data['Body'] = data.Body\
    .apply(lambda x : [word for word in x
                       if word not in stop_words])

data['Title'] = data.Title\
    .apply(lambda x : [word for word in x
                       if word not in stop_words])
```

```python
from nltk.stem.wordnet import WordNetLemmatizer

wn = WordNetLemmatizer()

data['Body'] = data.Body\
    .apply(lambda x : [wn.lemmatize(word) for word in x])

wn = WordNetLemmatizer()
data['Title'] = data.Title\
    .apply(lambda x : [wn.lemmatize(word) for word in x])
```
